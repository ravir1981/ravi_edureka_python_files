## Create Classes in python

1. to define class we create with `class` keyword
2. Variables inside a class is called `attribute`. To access an attribute, we call as - `object_name.attribute`
3. Functions defined inside a class is called as `methods`
4. class methods must have `self` as its first parameter (`self` is general name we use, but you are free to use any name)
5. By using `self` in the method, we pass object itself as a first paramter

````python
class learning:
  fruit = 'Apple' # Attribute of a class

  def display_msg(self):
    return "Good morning"
````

Create an object of a class:

`obj_name = class_name()`

````python
student1 = learning() # create an object
print(student1)
````

````python
student1.fruit
````

````python
student1.display_msg()
````

# Initializion of Class object

we use ``__init__()``

````python
class example:
  def __init__(self, name):
    self.name = name

  def message(self):
    return "Happy birthday "+self.name
````

````python
obj1 = example('John')
````

````python
obj1 = example("Jacob")
````

````python
obj1.name
````

````python
obj1.message()
````

### Steps :


1.   System searches for class definition of `example`
2.   creates a new object 
3.   calls the object's `__init__()` method by sending the newly created object as first argument, and the name `Jacob` as second argument
4.   Stores the value of `name` in the object
5.    Returns a new object
6. Attaches the variable `obj1` to the object


````python
class example:
  def __init__(zzzz, name):
    zzzz.name = name

  def message(zzzz):
    return "Happy birthday "+zzzz.name
````

````python
class example():
  def __init__(zzzz, name):
    zzzz.name = name

  def message(zzzz):
    return "Happy birthday "+zzzz.name
````

````python
# Inherit from Parent

class Car:
  def wheels(self):
    print("I am FWD car")

class SUV(Car):
  pass
````

````python
c1 = Car()
c2 = SUV()
````

````python
c1.wheels()
````

````python
c2.wheels()
````

````python
# Override a Method

class Car:
  def wheels(self):
    print("I am FWD car")

class SUV(Car):
  def wheels(self):
    print("I am a 4WD Car")
````

````python
c1 = Car()
c2 = SUV()
````

````python
c1.wheels()
````

````python
c2.wheels()
````

````python
# Adding a new method

class Car:
  def wheels(self):
    print("I am FWD car")

class SUV(Car):
  def wheels(self,name):
    self.name = name
    print("I am a 4WD Car ",self.name)

  def seater(self):
    print("This is a 6 seater car")
````

````python
c1 = Car()
c2 = SUV()
````

````python
c1.wheels()
````

````python
c2.wheels("Safari")
````

````python
c2.seater()
````

````python
c1.seater()
````
