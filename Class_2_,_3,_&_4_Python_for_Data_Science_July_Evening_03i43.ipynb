### Activity:
A school has following rules for grading system: - Below 25 - F - 25 to 45 - E - 45 to 50 - D - 50 to 60 - C - 60 to 80 - B - Above 80 - A

Ask user to enter marks and print the corresponding grade.

````python
num1=int(input("Enter the mark:"))
if num1 < 25:
    print("Student is FAIL", num1)
elif (num1 >= 25 and num1 <= 45):
    print("Student got score E",num1)
elif (num1 >= 45 and num1 <= 50):
    print("Student got score D",num1)
elif (num1 >= 50 and num1 <= 60):
    print("Student got score C",num1)
elif (num1 >= 60 and num1 <= 80):
    print("Student got score B",num1)
else:
    print("Student got score A",num1)

````
